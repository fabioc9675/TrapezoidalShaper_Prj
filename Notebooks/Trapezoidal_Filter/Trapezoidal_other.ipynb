### Conformacion de las se単ales exponenciales de eventos

Tipicamente los eventos de interaccion nuclear generan una se単al del tipo exponencial la cual desea ser representada por el filtro trapezoidal, 

A continuacion se construira la se単al a ser evaluada

![image](images/img4.png)

```python
# Importacion de librerias
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.signal import lfilter, firwin, freqz, filtfilt
```

![image](images/img6.jpg)

- Ecuacion en el continuo
$$X_{amp} = \frac{Q}{Cf}e^{-\frac{t}{\tau}}u(t)$$

- Ecuacion discreta
$$X_{amp} = \frac{Q}{Cf}\frac{1}{1-d z^{-1}}$$

```python
Q = 1  # Carga del detector
Cf = 1 # Capacitancia del detector
tao = 1/5 # constante de tiempo en segundos
Ts = 0.001 # Periodo de muestreo
N = 1/Ts

t = np.arange(0,10,Ts)
delay = 5
```

```python
def x_amp(A, delay):
    return A*(Q/Cf)*np.exp(-((t-delay)/tao))*np.heaviside(t-delay, 1)
```

```python
x = x_amp(1,delay)
```

```python
plt.figure(figsize=(10,3))
plt.plot(t,x)
plt.show
```

```python
x_noise = (x + 0.04*np.random.randn(len(t))) * 2**10

plt.figure(figsize=(10,3))
plt.plot(t,x_noise)
plt.show
```

---

## Calculo de la relacion senal a ruido

### Creacion de la senal limpia

```python
x_sim = x_amp(0.5,1) + x_amp(0.5,2) + x_amp(0.55,3) + x_amp(0.4,3.2) + x_amp(0.6,5) + x_amp(0.8,5.5)+ x_amp(0.4,6)+ x_amp(0.6,7)+ x_amp(0.5,8)+ x_amp(0.5,9)

plt.figure(figsize=(10,3))
plt.plot(t,x_sim)
plt.show
```

### Creacion de la senal con ruido

```python
x_noi = x_sim + 0.2* np.sin(100*np.pi*t)
x_noi = x_noi + 0.1* np.sin(0.2*np.pi*t)
x_noi = x_noi + 0.02*np.random.randn(len(t)) 
x_noi = x_noi + np.abs(np.min(x))

plt.figure(figsize=(10,3))
plt.plot(t,x_noi)
plt.show
```

### Separacion del ruido

```python
noise = (x_noi - x_sim)

plt.figure(figsize=(10,3))
plt.plot(t,noise)
plt.show
```

### Enhanced signal

```python
# Use firwin with a Kaiser window to create a lowpass FIR filter.
taps = firwin(200, [20/(2*N), 25/(2*N)], window='hann', pass_zero='bandpass')
#taps = firwin(100, [10/(2*N), 30/(2*N)], window='hann', pass_zero='bandpass')

# Use lfilter to filter x with the FIR filter.
x_filt = filtfilt(taps, 1.0, x_noi)

plt.figure(figsize=(10,3))
plt.plot(t,x_noi)
plt.plot(t,x_filt)
plt.show
```

### Calculo del ruido residual

```python
residual = x_filt - x_sim

plt.figure(figsize=(10,3))
plt.plot(t,residual)
plt.show
```

### Calculo de las SNR

```python
# Potencia de la se単al y del ruido
signal_power = np.mean(x_sim ** 2)
noise_power = np.mean(noise ** 2)
resid_power = np.mean(residual ** 2)

# Calcular SNR
snr_noise = 10 * np.log10(signal_power / noise_power)
snr_resid = 10 * np.log10(signal_power / resid_power)

print(f"SNR noise: {snr_noise:.2f} dB")
print(f"SNR filt: {snr_resid:.2f} dB")
print(f"SNR imp: {snr_resid-snr_noise:.2f} dB")
```

## Implementacion del filtro trapezoidal

```python
K = 500
L = 750
M = 2100
```

```python
def traps(x):
    y = np.zeros(len(x))
    b = np.zeros(len(x))
    for n in range((K+L),len(x)):
        A = x[n] - x[n-K] - x[n-L] + x[n-K-L]
        b[n] = b[n-1] + A
        C = b[n] + A*M
        
        y[n] = y[n-1] + C

    return y
```

```python
K = 29
L = 50
M = 200

y = traps(x_noise)
plt.figure(figsize=(10,3))
plt.plot(t,x_noise)
plt.plot(t,y)
plt.show
```

```python
# Use firwin with a Kaiser window to create a lowpass FIR filter.
#taps = firwin(100, [30/(2*N), 100/(2*N)], window='hann', pass_zero='bandpass')
taps = firwin(100, [10/(2*N), 30/(2*N)], window='hann', pass_zero='bandpass')

# Use lfilter to filter x with the FIR filter.
filtered_x = lfilter(taps, 1.0, data)

plt.figure(figsize=(10,3))
plt.plot(t,data)
plt.plot(t,filtered_x)
plt.show

coefs = np.round(taps * 2**10).astype(int)
df = pd.DataFrame(coefs)
df.to_csv('coefs.csv', index=False)
```

```python
K = 29
L = 50
M = 200

y = traps(filtered_x)
plt.figure(figsize=(10,3))
plt.plot(t,filtered_x)
plt.plot(t,y)
plt.show
```

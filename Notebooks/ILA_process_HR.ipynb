```python
# --- Cargar el vector desde el archivo CSV ---
filename = "alturas.csv"
loaded_df = pd.read_csv(filename)  # Leer el archivo CSV
loaded_vector = loaded_df["Values"].values  # Convertir la columna de nuevo a un array
print("Vector cargado desde el CSV:")
# plt.figure(figsize=(10,2.0))
#plt.plot(loaded_vector)

# Calcular la densidad (PDF) usando bins del histograma
bins = 800
count, bins = np.histogram(loaded_vector, bins=bins, density=True)
x = bins[:-1]  # Obtener los bordes izquierdos de los bins
y1 = 30*count  # Usar los valores de densidad del histograma

# Graficar línea con área sombreada
plt.figure(figsize=(10, 2.0))
plt.plot(x, y1, color='tab:blue', linewidth=1.0)
plt.fill_between(x, y1, color='skyblue', alpha=0.6)

# Configurar la leyenda y las etiquetas de los ejes
plt.xlabel('Channel')
plt.ylabel('Counts')
plt.xlim([5, 25])
plt.grid(True)

# Mostrar el gráfico
# plt.show()

plt.savefig("output/SimuHistogram.pdf", format="pdf", bbox_inches="tight")

```
